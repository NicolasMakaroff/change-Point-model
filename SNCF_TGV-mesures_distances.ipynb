```python
from scipy.stats import moment

def melange_gaussien(loi1,loi2,p1,p2):
    res = []
    for i in range(len(loi1)):
        res.append(p1*loi1[i]+p2*loi2[i])
    return res

def normal(x, mu_, sigma_):
    return (1/np.sqrt(2*np.pi*sigma_**2)) * np.exp(-1/2 * (x-mu_)**2/sigma_**2)


def Epsilon(S1,S2,a):
    m = len(S1)
    n = len(S2)
    
    distances = np.eye(m,n)
    for i in range(m):
        for j in range(n):
            distances[i,j] = abs(S1[i] - S2[j])**a
            
    distances2 = distances
    distances3 = distances
    
    for i in range(m):
        for j in range(n):
            if(not(i <= j <= m)):
                distances2[i,j] = 0
            if(not(i <= j <= n)):
                distances3[i,j] = 0
    
    return((2/(n*m)) * np.sum(distances) - (scipy.special.comb(n,2)**-1)*(np.sum(distances2)) - (scipy.special.comb(m,2)**-1)*(np.sum(distances3)))

    
def Q(S1,S2,a):
    m = len(S1)
    n = len(S2)
    return(Epsilon(S1,S2,a)*(m*n)/(n+m))

    
def rupture(X):
    stat_0 = []
    stat_1 = []
    stat_2 = []
    n = len(X)
    for k in range(1,n-2):
        S1 = X[0:k]
        S2 = X[k+1:n-1]
        n1 = len(S1)
        n2 = len(S2)
        if(n1 >= 2 and n2 >= 2):
            stat_0.append(Q(S1,S2,0))
            stat_1.append(Q(S1,S2,1))
            stat_2.append(Q(S1,S2,2))
        
    kmax_0 = None
    kmax_1 = None
    kmax_2 = None
    
    if(len(stat_0) != 0):
        kmax_0 = (np.where(stat_0 == max(stat_0))[0])[0]
    if(len(stat_1) != 0):
        kmax_1 = (np.where(stat_1 == max(stat_1))[0])[0]
    if(len(stat_2) != 0):
        kmax_2 = (np.where(stat_2 == max(stat_2))[0])[0]
    
    
    
    return(kmax_0, kmax_1, kmax_2)


#Pour obtenir plusieurs ruptures
def ruptures(X, m, res):
    
    res = sorted(res)
    n = len(X)
    List_X = []
    if(m == 0):
        return(list(set(res)))
    else:  
        for k in range(len(res)) :
            if(k == 0):
                List_X.append(X[0:res[k]])
            else:
                List_X.append(X[res[k-1]:res[k]])
        

        if(len(List_X)==0):
            r1 = rupture(X)[0]
            res.append(r1)
        else:
            for x in List_X:
                r_x = rupture(x)[0]
                if(len(x) >= 4 and r_x != None):
                    res.append(r_x)
            
        res = ruptures(X, m-1, res)
        return(list(set(res)))
    
    
```

```python
def Etat(gare,nom, colonne):
    rupt = rupture(gare[colonne].to_numpy())
    rupts = ruptures(gare[colonne].to_numpy(),6,[])
    print("rupts",rupts)
    #P1 = [normal(j,rupt[1],rupt[3]) for j in range(int(max(gare[colonne].values)+1))]
    #P2 = [normal(j,rupt[2],rupt[3]) for j in range(int(max(gare[colonne].values)+1))]
    print(rupt)
    print(gare[colonne].shape[0])
    p1 = rupt[0]/gare[colonne].shape[0]
    print('p1:{}'.format(p1))
    p2 = 1 - p1
    print('p2:{}'.format(p2))
    
    print("k_rupture = ", rupt[0])
    print(nom, " : ", colonne)
    fig = plt.figure(figsize=(10,7))
    ax1 = fig.add_subplot(111)
    # cacher les cadres : haut et droit
    ax1.spines['right'].set_visible(False)
    ax1.spines['top'].set_visible(False)


    ax1.xaxis.set_ticks_position('bottom')

    l1 = [k for k in range(rupt[0])]
    l2 = [k for k in range(rupt[0],len(gare[colonne].values))]
    
    #plt.plot(l1,gare[colonne].values[:rupt[0]], color = 'green')
    #plt.plot(l2,gare[colonne].values[rupt[0]:], color = 'blue')
    #plt.axvline(x=rupt[0],color='red')
    
    nr = len(rupts)
    for r in range(nr):
        if(r%2 == 0):
            col='green'
            col2='blue'
        else:
            col='blue'
            col2='green'
        
        if(r == 0):
            l1 = [k for k in range(rupts[r])]
            plt.plot(l1,gare[colonne].values[:rupts[0]], color = col)
            plt.axvline(x=rupts[0],color='red')
        elif(r == nr-1):
            l1 = [k for k in range(rupts[r-1],rupts[r])]
            l2 = [k for k in range(rupts[r],len(gare[colonne].values))]
            
            plt.plot(l1,gare[colonne].values[rupts[r-1]:rupts[r]], color = col)
            plt.plot(l2,gare[colonne].values[rupts[r]:], color = col2)
            plt.axvline(x=rupts[r],color='red')
        else:
            l1 = [k for k in range(rupts[r-1],rupts[r])]
            plt.plot(l1,gare[colonne].values[rupts[r-1]:rupts[r]], color = col2)
            plt.axvline(x=rupts[r],color='red')
    
    
    plt.show()

    print(nom, " : ", colonne)
    fig = plt.figure(figsize=(10,7))
    ax1 = fig.add_subplot(111)

    # cacher les cadres : haut , droit et gauche
    ax1.spines['right'].set_visible(False)
    #ax1.spines['left'].set_visible(False)
    ax1.spines['top'].set_visible(False)


    ax1.xaxis.set_ticks_position('bottom')
    #ax1.set_yticks([])

    plt.hist(gare[colonne].values,color=np.random.rand(3),density =True)

    #plt.plot(P1,color='g')
    #plt.plot(P2,color='b')
    #plt.plot(p1*np.array(P1)+p2*np.array(P2), color ='r')
    plt.show()
```
